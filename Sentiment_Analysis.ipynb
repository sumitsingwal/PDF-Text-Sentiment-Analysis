```python
import pdfminer
from pdfminer import high_level
import nltk.data
import nltk
nltk.download('punkt')
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import fitz
```

```python
#load the pdf file
def loadData(fileName):
  local_pdf_filename = fileName
  page = [0]
  extracted_text = high_level.extract_text(local_pdf_filename, "", page)
  return preprocess(extracted_text)
```

```python
# preprocess the text extracted from pdf
def preprocess(text_data):
  tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')
  sentences = tokenizer.tokenize(text_data)
  #return pdfHighlight(sentences)
  return sentiment(sentences)
  
```

```python
# Analyse sentiment using VADER
def sentiment(lines):
  analyzer = SentimentIntensityAnalyzer()
  neutral = []
  low_positive = []
  high_positive = []
  low_negative = []
  high_negative = []

  for sentence in lines:
    extracted_sentiment = analyzer.polarity_scores(sentence)
    if (extracted_sentiment['compound'] >  -0.05 and extracted_sentiment['compound'] < 0.05):
      neutral.append(sentence)
    elif (extracted_sentiment['compound'] >  0.05 and extracted_sentiment['compound'] <= 0.50):
      low_positive.append(sentence)
    elif (extracted_sentiment['compound'] >  0.50 and extracted_sentiment['compound'] <= 1):
      high_positive.append(sentence)
    elif (extracted_sentiment['compound'] >  -0.50 and extracted_sentiment['compound'] < -0.05):
      low_negative.append(sentence)
    elif (extracted_sentiment['compound'] >=  -1 and extracted_sentiment['compound'] <= -0.50):
      high_negative.append(sentence)
    print("{:-<65} {}".format(sentence, str(extracted_sentiment)))

  return pdfHighlight(neutral, low_positive, high_positive, low_negative, high_negative)
```

```python
# Highlight the sentence according to there sentiment score and save as pdf
def pdfHighlight(neutral, low_positive, high_positive, low_negative, high_negative):
  doc = fitz.open("The 21st Century Epigenomic Revolution.pdf")

  for page in doc:
      text = neutral
      for x in text:
        text_instances = page.searchFor(x)
        for inst in text_instances:
            highlight = page.addHighlightAnnot(inst)
            highlight.set_colors({"stroke":(0.3, 0.3, 0.3)}) #Highlight with light grey
            highlight.update()
      
      text = low_positive
      for x in text:
        text_instances = page.searchFor(x)
        for inst in text_instances:
            highlight = page.addHighlightAnnot(inst)
            highlight.set_colors({"stroke":(0, 1, 0)}) # Highlight with light green
            highlight.update()

      text = high_positive
      for x in text:
        text_instances = page.searchFor(x)
        for inst in text_instances:
            highlight = page.addHighlightAnnot(inst)
            highlight.set_colors({"stroke":(0, 0.5, 0)}) # Highlight with dark green
            highlight.update()
      
      text = low_negative
      for x in text:
        text_instances = page.searchFor(x)
        for inst in text_instances:
            highlight = page.addHighlightAnnot(inst)
            highlight.set_colors({"stroke":(1, 0, 0)}) # Highlight with light red
            highlight.update()
      
      text = high_negative
      for x in text:
        text_instances = page.searchFor(x)
        for inst in text_instances:
            highlight = page.addHighlightAnnot(inst)
            highlight.set_colors({"stroke":(0.5, 0, 0)}) # Highlight with dark red
            highlight.update()



  doc.save("Sentiment_Automated_Highlighted.pdf", garbage=4, deflate=True, clean=True)
```
